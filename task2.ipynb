```python
df1['Date']=pd.to_datetime(df1.Date)
```

```python
df1['Month_no']=df1.Date.dt.month
```

```python
# 1.Find out their avarage rating on weekly basis keep this in a mind that they take two days of leave
df1.groupby(['Agent Name','Day_of_week'])['Average Rating'].mean()
```

```python
# 2.Total working days for each agents 
df1.groupby('Agent Name')['Day_of_week'].count()
```

```python
# 3. Total query that you hvae taken 
df1.groupby('Agent Name')['Total Chats'].sum()
```

```python
# 4. total Feedback that you have received
df1.groupby('Agent Name')['Total Feedback'].sum()
```

```python
# 5. a agent name who have average rating between 3.5 to 4 
df1[(df1['Average Rating'] >=3.5) & (df1['Average Rating'] <=4)]['Agent Name'].unique()
```

```python
# 6 . Agent name who have rating lesss then 3.5 
df1[(df1['Average Rating'] <3.5)]['Agent Name'].unique()
```

```python
# 7 . agent name who have rating more then 4.5 
df1[(df1['Average Rating'] >4.5)]['Agent Name'].unique()
```

```python
# 8 . how many feedaback agents have received more then 4.5 average
df1.groupby('Agent Name')['Total Feedback'].mean()>=4.5
```

```python
# 9 . average weekly response time for each agent 

def avgTimeInSec(t):
    hh, mm, ss = t.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)

avg_response=[]
for i in df1['Average Response Time']:
    avg_response.append(avgTimeInSec(i))
avg_response
df1['Avg_responce_in_sec']=avg_response
```

```python
df1
```

```python
df1.groupby('Agent Name')['Avg_responce_in_sec'].mean()
```

```python
df1.columns
```

```python
# 10 . average weekely resolution time for each agents
avg_resolution=[]
for i in df1['Average Resolution Time']:
    avg_resolution.append(avgTimeInSec(i))
df1['Avg_resolution_in_sec']=avg_resolution
df1
```

```python
df1.groupby('Agent Name')['Avg_resolution_in_sec'].mean()
```

```python
# 11 . list of all agents name 
list(df1['Agent Name'].unique())
```

```python
# 12 . percentage of chat on which they have received a feedback
df1.groupby('Agent Name')[['Total Chats','Total Feedback']].sum()/10
```
